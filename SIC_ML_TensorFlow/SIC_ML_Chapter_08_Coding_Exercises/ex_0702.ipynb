### 1. Linear regression:

```python
import tensorflow as tf
import numpy as np
```

#### 1.1. Data:

```python
# Training data.
# hours of study (X) vs test score (y).
study = np.array([ 3, 4.5, 6, 1.2, 2, 6.9, 6.7, 5.5])          # Explanatory variable: X
score = np.array([ 88, 85, 90, 80, 81, 92, 95, 90])            # Response variable: y
```

#### 1.2. Define Variables and Placeholders:

```python
b1 = tf.Variable(1.0)                # A constant initial value.
b0 = tf.Variable(1.0)                # A constant initial value.
```

```python
X = tf.Variable(study, dtype=tf.float32)
y = tf.Variable(score, dtype=tf.float32)
```

#### 1.3. Define the model:

```python
def model(X):                                                   # Simple linear regression model.
  return b0 + b1 * X
```

#### 1.4. Define the loss function and the optimization method:

```python
def loss_fn(y_true, y_pred):
  return tf.reduce_sum(tf.square(y_true - y_pred))              # L2 loss.
```

```python
optimizer = tf.optimizers.SGD(learning_rate=0.001)
# optimizer = tf.optimizers.SGD(learning_rate=0.001, use_ema=True, ema_momentum=0.9)   # Momentum optimizer.
```

#### 1.5. Training and Testing:

```python
n_epochs = 5000                                                   # N# of epochs (gradient descent steps).
@tf.function
def train_step(X, y):
    with tf.GradientTape() as tape:
        y_pred = model(X)
        loss = loss_fn(y, y_pred)
    gradients = tape.gradient(loss, [b1, b0])
    optimizer.apply_gradients(zip(gradients, [b1, b0]))

# Training.
for i in range(n_epochs):
  train_step(X, y)
b0_model = b0.numpy()                                             # Get the final values of the Variables.
b1_model = b1.numpy()

# Testing.
mse_value = tf.reduce_mean(tf.square(y - model(X)))                  # Calculate the in-sample MSE.
```

#### 1.6. Display the result:

```python
print("Parameters b0 = {:5.3f} , b1 = {:5.3f}".format(b0_model, b1_model))
```

```python
print("MSE  = {:5.3f}".format(mse_value))
print("RMSE = {:5.3f}".format(np.sqrt(mse_value)))
```

#### 1.7. Prediction:

```python
# Define the testing data.
study_new = np.array([2.5, 3.3, 4.2]).reshape(-1,1)
```

```python
X = tf.Variable(study_new, dtype=tf.float32)
def model(X):                                                   # Define the prediction model.
  return b0_model + b1_model * X
```

```python
y_pred_value = model(X)
```

```python
# Predicted y values.
print(y_pred_value)
```
