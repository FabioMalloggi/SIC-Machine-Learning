Fancy indexing:

```python
arr = np.arange(100)
arrMask = ( (arr % 5) == 0 )
print(arr, "\n")
print(arrMask, "\n")
print(arr[arrMask], "\n")
print(type(arr[arrMask]))
```

```python
arrMask = ( ( (arr % 5) == 0 )  & ( arr >0) )
arr[arrMask]
```

Change the shape with reshape() method:

```python
a = np.arange(15)
a.reshape(3,5)
```

```python
# reshape() just produced a 'view'.
# So, there is no permanent change to 'a'.
a
```

```python
a = a.reshape(3,5)
a
```

```python
a = np.array([2,5,1,3])
a.shape
```

```python
a = a.reshape((2,2))
a
```

Careful that the reshape() method does not create a new object!

```python
a = np.arange(10)
b = a.reshape(2,5)
a[0] = -999
```

```python
b
```

```python
# Make a copy in order to create a new object.
c = a.reshape(2,5).copy()
c[0,0] = 0
a
```

#### 1.5. Modifying NumPy arrays:

Appending elements, rows and columns to arrays:

```python
a = np.array([1, 2, 3])
b = np.append(a, [4, 5, 6])
b
```

```python
a = np.array([[1, 2], [3, 4]])
b = np.append(a, [[9, 9]], axis=0)
print(a, "\n")
print(b)
```

```python
c = np.append(a, [[9], [9]] , axis=1)
c
```

```python
g = np.array([[9], [9]])
print(g)
print(g.shape, "\n")

d = np.array([[1, 2]])
print(d)
print(d.shape, "\n")

e = np.array([1, 2])
print(e)
print(e.shape, "\n")
print(e.ndim)
```

```python
d = np.array([[[[[[1, 2]]]]]])
print(d)
print(d.shape, "\n")
```
