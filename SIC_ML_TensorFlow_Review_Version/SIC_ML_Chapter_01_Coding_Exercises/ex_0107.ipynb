### 1. Class:

#### 1.1. Constructor method:

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print('A Dog object is created!')
```

```python
a_dog = Dog('Fido', 2)
```

#### 1.2. Destructor method:

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print('A Dog object is created!')
    def __del__(self):
        print('A Dog object is deleted!')
```

```python
a_dog = Dog('Fido', 2)
```

```python
del a_dog
```

#### 1.3. Memeber variable:

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
```

```python
dog1 = Dog('Fido', 2)
dog2 = Dog('Dido', 3)
```

```python
print("Name: {0} \nAge: {1}".format(dog1.name, dog1.age))
```

```python
print("Name: {0} \nAge: {1}".format(dog2.name, dog2.age))
```

#### 1.4. Class variable:

```python
class Dog:
    counter = 0                            # This is a class variable.
    def __init__(self, name):
        self.name = name
        Dog.counter += 1
    def __del__(self):
        Dog.counter -= 1
```

```python
# Create 2 Dog objects.
dog1 = Dog('Fido')
dog2 = Dog('Dido')
```

```python
Dog.counter
```

```python
del dog1
Dog.counter
```

```python
del dog2
Dog.counter
```

#### 1.5. Member method:

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def bark(self):
        print(self.name + ' is barking... woof... woof...')
```

```python
a_dog = Dog('Baduk', 2)
a_dog.bark()
```
